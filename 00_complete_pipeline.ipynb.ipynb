## Loading Products Data from Python to MySQL

````python
products = product[["product_upc",
                     "product_name",
                     "size",
                     "unit", "regular_price", "unit_cost",
                     "pack_size","department_ID", "vendor_ID", "brand_ID", "category_ID"]].drop_duplicates()
````

````python
products.to_sql(
    "product_data",
    con=engine,
    if_exists = "append",
    index = False
)
````

````python
department_clean = product[["department_ID",
                            "department_name"]].drop_duplicates()
````

````python
department_clean.to_sql(
    "department_data",
    con=engine,
    if_exists = "append",
    index = False
)
````

````python
category_data = product[["category_ID",
                         "category_name"]].drop_duplicates()
````

promotion data sending to sql

````python
promo = promotions[["promo_id", "product_upc", 
                         "promo_type", "discount_percent",
                         "start_date", "end_date"]].drop_duplicates()
````

````python
missing_upc = set(promotions["product_upc"]) - set(product["product_upc"])

missing_product = (
    promotions[promotions["product_upc"].isin(missing_upc)]
    [["product_upc", "product_name", "department_name", "brand"]].drop_duplicates()
)

missing_product["department_ID"] = (
    missing_product["department_name"].map(map_department)
)
missing_product["brand_ID"] = (
    missing_product["brand"].map(map_brand)
    )

missing_product.head()
````

````python
missing = missing_product[["product_upc", "product_name", "department_ID",
                           "brand_ID"]].drop_duplicates()
````

````python
missing.to_sql(
    "product_data",
    con=engine,
    if_exists = "append",
    index = False
)
````

````python
promo.to_sql(
    "promotion_data",
    con=engine,
    if_exists = "append",
    index = False
)
````
